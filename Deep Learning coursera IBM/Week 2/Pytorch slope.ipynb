# Gradient Descent the Hard way

```python
import torch


w = torch.tensor(-10.0, requires_grad = True)

X = torch.arange(-3, 3, 0.1).view(-1, 1) # does the same thing as range in py, range and step. so we make a range of X vals
f = -3*X 

import matplotlib.pyplot as plt

plt.plot(X.numpy(), f.numpy())
plt.show()

# now we will add some noise and plot that over the line

Y = f + 0.1 * torch.randn(X.size())

plt.plot(X.numpy(), Y.numpy(), 'ro')
plt.show()
```
