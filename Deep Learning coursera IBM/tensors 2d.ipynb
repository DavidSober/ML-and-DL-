```python
import torch
import numpy as np
import pandas as pd
```

```python
# making a 2D tensor first make a normal list 
a = [[11,12,13], [21,22,23], [31,32,33]]
# here is how you can cast it 
A = torch.tensor(a)
print(torch.tensor(a))

# ndimension obtains the dimensions ALSO REFERRED TO AS RANK
A.ndimension()
print(A.ndimension())

# to get all dimensions use shape
A.shape
print(A.shape)
```

```python
# numel method: gives you the number of elements name says it all
A.numel()
print(A.numel()) 
```

## Slicing and indexing 

```python
# indexing is EXACTLY the same as with a 2d python list

# Sliceing is much better with tensors when compared to normal lists
# for a simple slice of a couple row elements in py lists you do the following
a = [[11,12,13], [21,22,23], [31,32,33]]
pyslice = a[0][0:2] # we slice the first two elements from the first row
print(pyslice)
# However there is no analog way to slice a column in py. To do that you would need to do list comprehension 

# Tensors are much more simple 
A = torch.tensor(a)
# now let us slice the row as we did with the list above
tensorSlice = A[0,0:2] # note we use a comma separated list to specify the row num and col slice
print(tensorSlice)

# now let us do a col slice (which requires list comp in py)
tensorColSlice = A[1:3, 2]
print(tensorColSlice) # note we sliced the last col for the 2nd and 3rd row so easily
```

## Basic Operations in 2D

```python
X = torch.tensor([[1,0], [0,1]])
Y = torch.tensor([[2,1], [1,2]])
# element addition
Z = X + Y 
print(Z)
# scalar multiplication
print(Z * 2)

# element wise product
print('here is X: ')
print(X)
print('here is Y: ')
print(Y)
print('here is X * Y = ')
print(X * Y)

# matrix multiplication below
```

```python
# matrix multiplication
A = torch.tensor([[0,1,1], [1,0,1]])
B = torch.tensor([[1,1], [1,1], [-1,1]])

# MATRIX MULTIPLICATION
# recall rows must equal columns 
C = torch.mm(A, B)
print(C)

```
